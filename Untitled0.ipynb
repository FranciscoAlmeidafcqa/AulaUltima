# primeiro web scrapping
- Usando Python para buscar preços de produtos  no Mercado Livre

```python
import requests
from bs4 import BeautifulSoup
```

```python
url_site ='https://lista.mercadolivre.com.br/'
```

```python
produto= input('Digite o nome do produto')
print(produto)
```

```python
produto_tratado =produto.replace(' ', '-')
print(produto_tratado)
```

```python
link_produtos= url_site + produto_tratado
```

```python
print(link_produtos)
```

```python
# ver se o site aceita a requisição(conteudo do site - pagina em texto html)
response =requests.get(link_produtos)
```

```python
print(response)
```

```python
#html.parse  é o analizador de html que iremos  utilizar para analisar o conteudo  do site
site = BeautifulSoup(response.text, "html.parser")
```

```python
# find_all()= metodo da classe BeautifulSoup   que permite  buscar tosas as ocorrencisa  de uma tg HTML especifica
dados = site.find_all("div", attrs = {"class" : "ui-search-result__content-wrapper shops__result-content-wrapper"})
```

```python
# filtro no site  em geral para  apenas a div  com a classe que quero buscar as inforações
print(dados)
```

```python
nomes = []
preco = []

for x in dados:
  #nome do produto
  nome_produto = x.find_all("h2")[0].text

  #preço
  preco_produto = x.find_all("span", class_="andes-money-amount__fraction")[0].text

  nomes.append(nome_produto)
  preco.append(preco_produto)
```

```python
len(nomes)
```

```python
import pandas as pd
```

```python
dados_coletados={
    "Nomes":nomes,
    "Preco" :preco
}
```

```python
pd.DataFrame(dados_coletados)
```
